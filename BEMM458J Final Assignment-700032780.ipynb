## Task 1. Which Twitter users are the most popular ? (10%)
### For each Task please: (1) develop the code required and (2) provide a brief discussion and interpretation of the results

### Tip: consider retweet and favorite counts as proxies for popularity

```python
Artist=ConversationsLean[ConversationsLean['stakeholder']== 'Artist']
Multinational=ConversationsLean[ConversationsLean['stakeholder']== 'Multinational']
NGO=ConversationsLean[ConversationsLean['stakeholder']== 'NGO']
OtherInstitution=ConversationsLean[ConversationsLean['stakeholder']== 'OtherInstitution']
Scientific=ConversationsLean[ConversationsLean['stakeholder']== 'Scientific']
```

```python
import numpy as np
artist_favourite=np.sum(Artist['tweet.user.favourites_count'])
Multinational_favourite=np.sum(Multinational['tweet.user.favourites_count'])
NGO_favourite=np.sum(NGO['tweet.user.favourites_count'])
OtherInstitution_favourite=np.sum(OtherInstitution['tweet.user.favourites_count'])
Scientific_favourite=np.sum(Scientific['tweet.user.favourites_count'])
print('The tweet favourite number of Artist is',artist_favourite)
print('The tweet favourite number of Multinational is',Multinational_favourite)
print('The tweet favourite number of NGO is',NGO_favourite )
print('The tweet favourite number of OtherInstitution is',OtherInstitution_favourite)
print('The tweet favourite number of Scientific is',Scientific_favourite )
```

```python
artist_retweet=np.sum(Artist['tweet.retweet_count'])
Multinational_retweet=np.sum(Multinational['tweet.retweet_count'])
NGO_retweet=np.sum(NGO['tweet.retweet_count'])
OtherInstitution_retweet=np.sum(OtherInstitution['tweet.retweet_count'])
Scientific_retweet=np.sum(Scientific['tweet.retweet_count'])
print('The retweet number of Artist is',artist_retweet)
print('The retweet number of Multinational is',Multinational_retweet)
print('The retweet number of NGO is',NGO_retweet )
print('The retweet number of OtherInstitution is',OtherInstitution_retweet)
print('The retweet number of Scientific is',Scientific_retweet )
```

```python
popular=ConversationsLean[['stakeholder','tweet.retweet_count','tweet.user.favourites_count']]
```

```python
#use pivot to compare better
pd.pivot_table(data=popular,index='stakeholder',values=['tweet.retweet_count','tweet.user.favourites_count'],aggfunc=np.sum)
```

After summarsing retweets number and faourite tweets number, we can find that NGO's tweets have been forwarded and liked the most.

## Task 3. Do emotions play a role in the number of retweets and favorites achieved by tweets ? (20 %)
### For each Task please: (1) develop the code required and (2) provide a brief discussion and interpretation of the results

### Tip:  correlations between variables
### Tip:  visualizations relating variables
### Tip:  optionally consider basic regression models to determine the impact of some variables on others (e.g. impact of love on favorite, impact of fear on retweet)

```python
corr_df=ConversationsLean[['anger','fear','joy','love','sadness','trust','identity_hate','insult','obscene','severe_toxic','threat','toxic','tweet.retweet_count','tweet.favorite_count', 'stakeholder']]
corr_df
```

```python
#calculate correaltion among all variables
corrs=corr_df.corr(method='pearson')
corrs
```

```python
#heatmap covariance

plt.subplots(figsize=(10, 8))
sns.heatmap(corrs, cmap="YlGnBu",annot=True)
plt.title('Correlation matrix')
plt.style.use("dark_background")

```

From the heatmap, we could find that both retweet_count and favoutite_count is positive related to anger, fear and sadness.

And they have a negative relationship with joy, love, trust, insult toxic threat and obscene.
